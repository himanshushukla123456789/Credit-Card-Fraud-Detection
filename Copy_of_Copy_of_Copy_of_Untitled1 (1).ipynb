**MODEL BUILDING**

**Logistic Regression With Random Undersampling**

```python
from collections import Counter
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score,recall_score,confusion_matrix,accuracy_score,f1_score,classification_report
```

```python
from imblearn.under_sampling import RandomUnderSampler
under_sample=RandomUnderSampler(random_state=5)
X_resampled_us,y_resampled_us=under_sample.fit_resample(X_train,y_train)
```

```python
print(sorted(Counter(y_resampled_us).items()))
```

```python
lreg_us=LogisticRegression()
lreg_us.fit(X_resampled_us,y_resampled_us)
y_train_pred=lreg_us.predict(X_resampled_us)
y_test_pred=lreg_us.predict(X_test)
```

```python
model_name='Logistic Regression With Random UnderSampling'
train_accuracy=accuracy_score(y_resampled_us,y_train_pred)
test_accuracy=accuracy_score(y_test,y_test_pred)
Precision=precision_score(y_test,y_test_pred)
Recall=recall_score(y_test,y_test_pred)
fscore=f1_score(y_test,y_test_pred)
model_data=[[model_name,train_accuracy,test_accuracy,Precision,Recall,fscore]]
Score_df=pd.DataFrame(model_data)
Score_df.columns=['Model_Name','Train Accuracy','Test Accuracy','Precision','Recall','F-1 Score']
Score_df
```

**Logistic Regression with OverSampling**

```python
from imblearn.over_sampling import RandomOverSampler
over_sample=RandomOverSampler(sampling_strategy=1)
X_resampled_os,y_resampled_os=over_sample.fit_resample(X_train,y_train)
len(X_resampled_os)
```

```python
print(sorted(Counter(y_resampled_os).items()))
```

```python
lreg_os=LogisticRegression()
lreg_os.fit(X_resampled_os,y_resampled_os)
y_train_pred=lreg_os.predict(X_resampled_os)
y_test_pred=lreg_os.predict(X_test)
```

```python
model_name='Logistic Regression With Random OverSampling'
train_accuracy=accuracy_score(y_resampled_os,y_train_pred)
test_accuracy=accuracy_score(y_test,y_test_pred)
Precision=precision_score(y_test,y_test_pred)
Recall=recall_score(y_test,y_test_pred)
fscore=f1_score(y_test,y_test_pred)
model_data=[[model_name,train_accuracy,test_accuracy,Precision,Recall,fscore]]
Score_df_1=pd.DataFrame(model_data)
Score_df_1.columns=['Model_Name','Train Accuracy','Test Accuracy','Precision','Recall','F-1 Score']
Score_df_2=pd.concat([Score_df,Score_df_1],ignore_index=True)
Score_df_2.head()
```
